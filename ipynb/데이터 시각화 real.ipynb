```python
import matplotlib.pyplot as plt

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'Age': [25, 30, 35, 40, 10]}

# bar plot
plt.bar(data['Name'], data['Age'])

plt.xlabel('Name')
plt.ylabel('Age')
plt.title('People Ages')

# %matplotlib inline 설정이 디폴트이기에, plt.show()없이도 차트가 보여집니다.
# plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'Age': [25, 30, 35, 40, 10]}

# Bar plot
sns.barplot(x='Name', y='Age', data=data)

plt.xlabel('Name')
plt.ylabel('Age')
plt.title('People Ages')
```

```python
from bokeh.plotting import figure, show
#from bokeh.io import output_notebook
#from bokeh.models import ColumnDataSource

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'Age': [25, 30, 35, 40, 10]}

# Create a Bokeh figure
output_notebook()
p = figure(
    x_range=data['Name'],
    height=350,
    title='People Ages',
    x_axis_label='Name',
    y_axis_label='Age',
)

# Convert data to Bokeh ColumnDataSource
source = ColumnDataSource(data=data)

# Bar plot
p.vbar(x='Name', top='Age', width=0.5, source=source)

# Customize plot aesthetics
p.xaxis.major_label_orientation = 1
p.title.text_font_size = '16px'
p.yaxis.axis_label_standoff = 12

# Show the Bokeh plot
show(p)
```
